## 1. Remove all Null Values in Numerical Columns

```python
# First checking the total null values in each numerical column
for i in Fiza.columns:
    if Fiza[i].dtype== "float64" and"int64":
        print(i,Fiza[i].isnull().sum())
```

```python
# Now replacing that null values with maximum
for i in Fiza.columns:
    if (Fiza[i].dtype=="float64" and "int64"):
        if (Fiza[i].isnull().sum()>0):
            Fiza[i]=Fiza[i].fillna(Fiza[i].max())
Fiza
```

```python
# Now checking either the null values left in any numerical column or not?
for i in Fiza.columns:
    if Fiza[i].dtype=="float64" and "int64":
        print(i,Fiza[i].isnull().sum())
```

```python
# All null values replaced successfully
```

## 2.Convert all 0 values in Numerical Columns with 1.


```python
for i in Fiza.columns:
    if Fiza[i].dtype=="float64" and "int64":
        Fiza[i]=Fiza[i].replace(0,1)
```
